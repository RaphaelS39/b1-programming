```python
import math

def f(x):
    return math.log(x+1) + x - 2

def bisection(f, a, b, tol=1e-6, max_iter=1000):
    fa = f(a)
    fb = f(b)
    if fa * fb > 0:
        raise ValueError("f(a) und f(b) mÃ¼ssen unterschiedliche Vorzeichen haben.")
    it = 0
    while (b - a) / 2.0 > tol and it < max_iter:
        m = 0.5 * (a + b)
        fm = f(m)
        if fa * fm <= 0:
            b = m
            fb = fm
        else:
            a = m
            fa = fm
        it += 1
    m = 0.5 * (a + b)
    return m, it
```
