```python
###Exercise 1: Failed login detctor

login_attempts = [("alice", "success"),("bob", "failed"),("bob", "failed"),("charlie", "success"),("bob", "failed"),("alice", "failed")]

print("Checking login attempts...")

failed_counts = {}




for username, status in login_attempts:
    if status == "failed":
        if username in failed_counts:
            failed_counts[username] += 1
        else:
            failed_counts[username] = 1

for user,counts in failed_counts.items():
    if counts >= 3 :
        print(f"ALERT: User '{user}' has {counts} failed login attempts")

print('Security check complete')    
    
            
    
    
```

```python
###Exercise 2: Port security scanner
devices = [ ("192.168.1.10", [22, 80, 443]),("192.168.1.11", [21, 22, 80]), ("192.168.1.12", [23,80, 3389])]
risky_ports = [21, 23, 3389]

print('Scanning network devices ...')

total_risks = 0

for ip,ports in devices:

    for port in ports:

        if port in risky_ports:
            print(f"WARNING: {ip} has risky port {port} open")
            total_risks +=1

print(f"Scan complete: {total_risks} security risks found")



```

```python
###Exercise 3: Password policy validator

passwords = [ "Pass123", "SecurePassword1", "weak", "MyP@ssw0rd", "NOLOWER123"]

compliant_count = 0
noncompliant_count = 0

for pw in passwords:

    reasons = []
    
    if len(pw) < 8:
        reasons.append('Too short')#


    has_upper = False
    for char in pw:
        if char.isupper():
            has_upper = True
            break
    if not has_upper:
            reasons.append('No uppercase')


    has_lower = False
    for char in pw:
        if char.islower():
            has_lower = True
            break
    if not has_lower:
            reasons.append('No lowercase')


    has_digit = False
    for char in pw:
        if char.isdigit():
            has_digit = True
            break
    if not has_lower:
            reasons.append('No digits')

    if len(reasons) == 0:
        print(f"PASS: '{pw}' - Meets all requirements")
        compliant_count += 1
    else:
        print(f"FAIL: '{pw}' - " + ", ".join(reasons))
        noncompliant_count += 1


print(f"Summary: {compliant_count} compliant, {noncompliant_count} non-compliant")




```
