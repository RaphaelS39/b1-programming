```python
#Part A

import string
import random

def check_min_length(password: str, min_len: int = 8) -> bool:
    """Prüft, ob password mindestens min_len Zeichen hat."""
    return len(password) >= min_len

def has_uppercase(password: str) -> bool:
    """Prüft, ob password mindestens einen Großbuchstaben enthält."""
    for ch in password:
        if ch.isupper():
            return True
    return False

def has_lowercase(password: str) -> bool:
    """Prüft, ob password mindestens einen Kleinbuchstaben enthält."""
    for ch in password:
        if ch.islower():
            return True
    return False

def has_digit(password: str) -> bool:
    """Prüft, ob password mindestens eine Ziffer enthält."""
    for ch in password:
        if ch.isdigit():
            return True
    return False

def has_special_char(password: str, special_chars: str = string.punctuation) -> bool:
    """
    Prüft, ob password mindestens ein Sonderzeichen enthält.
    Default special_chars ist string.punctuation (!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~).
    """
    for ch in password:
        if ch in special_chars:
            return True
    return False



#Part B

def validate_password(password: str) -> dict:
    """
    Ruft alle Einzelprüfungen auf und liefert ein Dictionary mit Ergebnissen.
    Struktur:
    """
    results = {
        'check_min_len': check_min_length(password),
        'has_upper': has_uppercase(password),
        'has_lower': has_lowercase(password),
        'has_digit': has_digit(password),
        'has_special_char': has_special_char(password)
    }

    
    results['is_valid'] = all(results.values())
    return results



#Part C

def give_hint(failed_checks: list) -> str:
    """
    Liefert eine kurze, zufällige Hilfenachricht basierend auf den fehlgeschlagenen Checks.
    (Zeigt nur einen Grund als Tipp, nicht alle.)
    """
    hints = {
        'check_min_len': "Versuche ein längeres Passwort (mindestens 8 Zeichen).",
        'has_upper': "Füge mindestens einen Großbuchstaben hinzu (z. B. 'A').",
        'has_lower': "Füge mindestens einen Kleinbuchstaben hinzu (z. B. 'a').",
        'has_digit': "Füge mindestens eine Zahl hinzu (z. B. '1').",
        'has_special_char': "Füge ein Sonderzeichen hinzu (z. B. '!','@','#')."
    }
    # Wähle zufällig einen der fehlgeschlagenen Checks und gib die passende Nachricht zurück
    choice = random.choice(failed_checks)
    return hints.get(choice, "Versuche ein stärkeres Passwort.")

def print_results(password: str, results: dict) -> None:
    """
    Gibt für jede Regel aus, ob sie erreicht wurde (einmal pro Regel).
    Am Ende: Gesamturteil und evtl. ein Tipp.
    """
    print("\nValidation results:")
    # Klar benannte Ausgaben für jede Regel
    print(f" - Minimum length (8): {'OK' if results['check_min_len'] else 'NOT MET'}")
    print(f" - Has uppercase:     {'OK' if results['has_upper'] else 'NOT MET'}")
    print(f" - Has lowercase:     {'OK' if results['has_lower'] else 'NOT MET'}")
    print(f" - Has digit:         {'OK' if results['has_digit'] else 'NOT MET'}")
    print(f" - Has special char:  {'OK' if results['has_special_char'] else 'NOT MET'}")

    # Gesamtbewertung
    if results['is_valid']:
        print("\nOverall: PASS — Strong password ")
    else:
        print("\nOverall: FAIL — Password does not meet all requirements ")
        # Liste der fehlgeschlagenen Checks (nur Schlüssel-Namen)
        failed = [k for k, v in results.items() if (k != 'is_valid') and (v is False)]
        # Gib einen zufälligen Tipp basierend auf einem der fehlenden Kriterien
        hint = give_hint(failed)
        print("Hint:", hint)

def main():
    # Einfaches UI: Passwort-Eingabe vom Nutzer
    print("Please enter a password to validate:")
    pwd = input("Password: ")

    # Validieren
    results = validate_password(pwd)

    # Ergebnisse ausgeben
    print_results(pwd, results)

if __name__ == "__main__":
    main()

```
